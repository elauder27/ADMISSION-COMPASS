```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

lasu_df = pd.read_csv("../Data/LASU.csv")

lasu_df.head()
lasu_df.info()
lasu_df.describe()
lasu_df.isnull().sum()
lasu_df.duplicated().sum()

print("Shape:", lasu_df.shape)
print(lasu_df["admitted"].value_counts())
print(lasu_df["screening_score"].describe())

# Screening histogram
plt.figure(figsize=(8,5))
sns.histplot(lasu_df["screening_score"], bins=30, kde=True)
plt.title("Screening Score Distribution (LASU)")
#plt.savefig("static/screening_hist.png") if os.path.exists("static") else plt.show()

```

```python
# Admission vs Screening boxplot
plt.figure(figsize=(8,5))
sns.boxplot(x="admitted", y="screening_score", data=lasu_df)
plt.title("Screening Score by Admission Label")
#plt.savefig("static/screening_by_admit.png") if os.path.exists("static") else plt.show()
```
