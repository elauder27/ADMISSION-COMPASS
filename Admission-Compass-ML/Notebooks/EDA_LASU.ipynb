```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

lasu_df = pd.read_csv("../Data/LASU.csv")

lasu_df.head()
lasu_df.info()
lasu_df.describe()
lasu_df.isnull().sum()
lasu_df.duplicated().sum()

sns.histplot(lasu_df['utme_score'], bins=30, kde=True)
plt.show()

sns.countplot(x="admitted", data=lasu_df)
plt.show()
```

```python
lasu_df.groupby("faculty")['admitted'].value_counts(normalize=True).unstack()

# Encode admitted column (Admitted=1, Rejected=0)
lasu_df['admitted_numeric'] = lasu_df['admitted'].map({"Admitted": 1, "Rejected": 0})

plt.figure(figsize=(8,6))
sns.heatmap(lasu_df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.show()
```
